```python
import turtle
t = turtle.Turtle()

# Bước 1: Vẽ nhà

t.penup()
t.bk(500)
t.pendown()

t.color("black", "blue")
t.begin_fill()
for i in range(4):
    t.fd(100)
    t.rt(90)
t.end_fill()

t.color("black", "magenta")
t.begin_fill()
t.lt(45)
t.fd(70.71)
t.rt(90)
t.fd(70.71)
t.end_fill()

t.penup()
t.bk(70.71)
t.pendown()

t.color("black", "orange")
t.begin_fill()
t.left(60)
t.fd(90)
t.rt(60)
t.fd(70.71)
t.rt(120)
t.fd(90)
t.end_fill()

t.lt(75)

t.penup()
t.fd(100)
t.pendown()

t.color("black", "yellow")
t.begin_fill()
t.lt(105)
t.fd(90)
t.lt(75)
t.fd(100)
t.end_fill()

t.penup()
t.bk(100)
t.pendown()

t.lt(105)
t.fd(30)
t.rt(105)
t.end_fill()

t.penup()
t.fd(66)
t.pendown()

t.color("black", "brown")
t.begin_fill()
for i in range(2):
    t.lt(105)
    t.fd(33)
    t.lt(75)
    t.fd(33)
t.end_fill()

t.penup()
t.fd(33)
t.lt(105)
t.fd(66)
t.rt(15)
t.fd(66)
t.lt(90)
t.fd(33)
t.pendown()

t.color("black", "palegreen")
t.begin_fill()
for i in range(2):
    t.fd(64)
    t.lt(90)
    t.fd(33)
    t.lt(90)
t.end_fill()
# chuyển sang vẽ mặt trời

t.lt(90)
t.penup()
t.fd(200)
t.lt(90)
t.fd(200)
t.pendown()


t.color("black", "yellow")
t.begin_fill()
for i in range(4):
    for i in range(1):
        t.lt(90)
        t.fd(30)
        t.penup()
        t.bk(30)
        t.rt(90)
        t.pendown()
    for i in range(45):
            t.rt(1)
            t.fd(0.5)
    t.lt(90)
    t.fd(15)
    t.bk(15)
    
    t.rt(90)
    for i in range(45):
        t.rt(1)
        t.fd(0.5)
t.end_fill()            
# chuyển sang vẽ cây
t.penup()
t.rt(90)
t.fd(100)
t.rt(90)
t.fd(100)
t.pendown()

t.color("black", "palegreen")
t.begin_fill()
for i in range(3):
    t.rt(45)
    t.fd(60)
    t.lt(135)
    t.fd(84.85)
    t.lt(135)
    t.fd(60)
    t.lt(135)
    t.penup()
    t.fd(42.43)
    t.pendown()
t.end_fill()            
    
t.rt(90)
t.penup()
t.fd(14.14)
t.pendown()

t.color("black", "brown")
t.begin_fill()
for i in range(2):
    t.lt(90)
    t.fd(31)
    t.lt(90)
    t.fd(21)
t.end_fill()            
```
